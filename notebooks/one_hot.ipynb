```python
list_of_genres = list(df['genres'].apply(lambda x: ' '.join(sorted(literal_eval(x)) if isinstance(x, str) else sorted(x))))
list_of_genres
```

```python
Counter(' '.join(list_of_genres).split()).most_common()
```

```python
top_genres = [m[0] for m in Counter(' '.join(list_of_genres).split()).most_common(15)]
df = df.drop(['genres'], axis=1)
```

```python
for g in top_genres:
    df['genre_' + g] = df['all_genres'].apply(lambda x: 1 if g in x else 0)
```

```python
df['keywords'].apply(lambda x: ' '.join(sorted(literal_eval(x)) if isinstance(x, str) else sorted(x)))
```

```python
list_of_keywords = []
for keywords in df['keywords']:
    if isinstance(keywords, str):
        parsed = ast.literal_eval(keywords)
        list_of_keywords.append(sorted(parsed))
    else:
        list_of_keywords.append(sorted(keywords))
list_of_keywords
```

```python
top_keywords = [m[0] for m in Counter([i for j in list_of_keywords for i in j]).most_common(30)]
```

```python
df['num_keywords'] = df['keywords'].apply(lambda x: len(x))
df['all_keywords'] = df['keywords'].apply(lambda x: ' '.join(sorted(literal_eval(x)) if isinstance(x, str) else sorted(x))) 
for g in top_keywords:
    df['keyword_' + g] = df['all_keywords'].apply(lambda x: 1 if g in x else 0)
```

```python
def get_names(x):
    if pd.isna(x):
        return []
    if isinstance(x, str):
        x = ast.literal_eval(x)  
    if isinstance(x, list):
        return [p.get('name') for p in x if isinstance(p, dict) and p.get('name')]
    return []
```

```python
cast_names = df['cast'].apply(get_names)
cast_names
```

```python
Counter(cast_names.explode().dropna()).most_common(15)
```

```python
df['cast_num'] = df['cast'].apply(lambda x: len(x) if x != {} else 0)
top_cast_name = list(Counter(cast_names.explode().dropna()).most_common(15))
for g in top_cast_name:
    df['cast_name_' + g[0]] = df['cast'].apply(lambda x: 1 if g[0] in str(x) else 0)
```

```python
df.head()
```

```python
top_directors = Counter(df['director'].dropna()).most_common(15)
for g in top_directors:
    df['director_name_' + g[0]] = df['director'].apply(lambda x: 1 if g[0] in str(x) else 0)
```
